### Email Spam detection


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,roc_auc_score,roc_curve
```

```python

df=pd.read_csv("emails.csv")
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,1:3001]
y=df.iloc[:,-1].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)

print("Train Result\n")
y_train_pred=knn.predict(x_train)
#y_train_prob=knn.predict_prob(x_train)[:,1]

print("confusion matrix:\n",confusion_matrix(y_train,y_train_pred))
print("Accuracy:",accuracy_score(y_train,y_train_pred))
#print("ROC AUC for Train :",roc_auc_score(y_train,y_train_prob))

print("Test Result\n")
y_test_pred=knn.predict(x_test)
#y_test_prob=knn.predict_prob(x_test)[:,1]
print("confusion matrix:\n",confusion_matrix(y_test,y_test_pred))
print("Accuracy:",accuracy_score(y_test,y_test_pred))
#print("ROC AUC for Train :",roc_auc_score(y_test,y_test_prob))
```

```python
from sklearn.svm import SVC
svm=SVC(probability=True)
svm.fit(x_train,y_train)

print("Train Result\n")
y_train_pred=svm.predict(x_train)
#y_train_prob=svm.predict_prob(x_train)[:,1]

print("confusion matrix:\n",confusion_matrix(y_train,y_train_pred))
print("Accuracy:",accuracy_score(y_train,y_train_pred))
#print("ROC AUC for Train :",roc_auc_score(y_train,y_train_prob))

print("Test Result\n")
y_test_pred=svm.predict(x_test)
#y_test_prob=svm.predict_prob(x_test)[:,1]
print("confusion matrix:\n",confusion_matrix(y_test,y_test_pred))
print("Accuracy:",accuracy_score(y_test,y_test_pred))
#print("ROC AUC for Train :",roc_auc_score(y_test,y_test_prob))
```
